```python
### DEFINE ###
#Make column names snake_case in house_data_copy

### CLEAN ### 
col_copy = house_data_copy.columns.copy()
corrected_col = []
for col in col_copy:
    if col.find(" "):
        col = col.replace(" ", "_")
    if col.find("?"): #took care of the puncation in this step also
        col = col.replace("?", "")
    
    corrected_col.append(col)

house_data_copy.columns = corrected_col
        


### TEST ###
house_data_copy.head()
```
